```python
anime_data.head()
```

```python
has_a_genre = anime_data[[type(x)==str for x in list(anime_data.genre)]]
has_a_genre
```

```python
all_genres = ""
for film in list(has_a_genre.genre):
    all_genres += film + ", "
genre_type = all_genres.split(",")
g = [x.strip() for x in genre_type]
genres = set (g)
print(genres)
```

```python
genres.remove('')
for g in genres:
    has_a_genre[g] = [g in x for x in has_a_genre.genre]
has_a_genre.head()
```
